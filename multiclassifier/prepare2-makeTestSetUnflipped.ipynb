```python
def sumRow(X):
    sum1 = 0
    sumList = []
    for i in X:
        sum1 = np.sum(i,axis=0)
        sumList.append(sum1)
        b = np.array(sumList)
    return b

list1 = []
for index, row in backToPandas.iterrows(): 
#for index, row in df1.iterrows():   
    rowSum = 0.0
    X = row.values
    X = np.reshape(X,(13,21))
    rowSum = sumRow(X)
    list1.append(rowSum)
#list1
```
